```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv('pulsar_stars.csv')
```

The first 10 rows of the dataset are,

```python
data.head(10)
```

The last 10 rows of the dataset are,

```python
data.tail(10)
```

The dimensions of the dataset are,

```python
print("The number of rows= {}".format(data.shape[0]))
print("The number of columns= {}".format(data.shape[1]))
```

```python
X = data.iloc[:, :8].values
y = data.iloc[:, 8].values
```

Check for missing data

```python
print (data.isnull().sum())
```

# Distribution of target variables

```python
color = ['#5FD9E1', '#B4F354']
plt.subplot(122)
plt.pie(data["target_class"].value_counts().values,
        labels=["No","Yes"], radius=1.2, autopct='%0.1f%%', explode=[0, 0.2], colors=color, shadow=True)
plt.title("Proportion of target variable in dataset")
plt.show()
```

# Pair Plot

```python
sns.pairplot(data,hue="target_class")
plt.title("Pair plot between the variables")
plt.show()
```

Split the dataset into training set and testing set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

Perform feature scaling on the matrix of independant variables

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# 1. K Neighbors Classification

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
classifier.fit(X_train, y_train)
```

Predict if it is a pulsar star or not

```python
y_pred = classifier.predict(X_test)
```

Confusion Matrix for K Neighbors Classification

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
cm = confusion_matrix(y_pred, y_test)
print("The confusion matrix is, \n{}".format(cm))
print("The accuracy is= {}".format(accuracy_score(y_pred, y_test)))
print("The precision is= {}".format(precision_score(y_pred, y_test)))
print("The recall is= {}".format(recall_score(y_pred, y_test)))
print("The f1 score= {}".format(f1_score(y_pred, y_test)))
```

The CAP curve is,

```python
total = len(y_test)
class_1_count = np.sum(y_test)
class_0_count = total - class_1_count
plt.figure(figsize=(20,12))
plt.plot([0, total], [0, class_1_count], c = '#0F30E5', linestyle = '--', label = 'Random Model')
plt.plot([0, class_1_count, total], 
         [0, class_1_count, class_1_count], 
         c = '#5AF37F', 
         linewidth = 2, 
         label = 'Perfect Model')
probs = classifier.predict_proba(X_test)
probs = probs[:, 1]
model_y = [y for _, y in sorted(zip(probs, y_test), reverse = True)]
y_values = np.append([0], np.cumsum(model_y))
x_values = np.arange(0, total + 1)
plt.plot(x_values, 
         y_values, 
         c = '#E75130', 
         label = 'K Neighbors Classifier', 
         linewidth = 4)

# Plot information
plt.xlabel('Total observations', fontsize = 16)
plt.ylabel('Is a pulsar star', fontsize = 16)
plt.title('Cumulative Accuracy Profile', fontsize = 16)
plt.legend(loc = 'lower right', fontsize = 16)
plt.show()
```

# Conclusion
It can be concluded that based on the confusion matrix, the accuracy of the predictions is 98.3% whereas the F1 score of the prediction is 88.92%. This is due to the fact that the classification is skewed which was interpreted from the pie chart constructed above.
By examining the CAP curve, we can deduce that the model is very accurate to the perfect model.

# Using Grid search method to determine the best hyper parameters

Construct a list of the hyper parameters used by the KNeighborsClassification class and add the possible values for the 1st iteration.

```python
from sklearn.model_selection import GridSearchCV
grid_params = {'n_neighbors' : [3, 5, 10, 20, 30],
               'weights' : ['uniform', 'distance'],
               'p' : [1,2]}
```

```python
import pandas as pd
data = pd.read_csv('pulsar_stars.csv')
X = data.iloc[:, :8].values
y = data.iloc[:, 8].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)
```

Execute the GridSearchCV function to determine the best parameters for the classification model.

```python
gs = GridSearchCV(estimator=classifier, param_grid=grid_params, verbose=1, cv=10, n_jobs=-1)
gs_results = gs.fit(X_train, y_train)
```

View the KNeighborsClassifier function to see the best estimator.

```python
gs_results.best_estimator_
```

Determine the best parameters for the model.

```python
gs_results.best_params_
```

Determine the score of the new obtained parameters that is fitted into the model.

```python
gs_results.best_score_
```

```python
import pandas as pd
data = pd.read_csv('pulsar_stars.csv')
X = data.iloc[:, :8].values
y = data.iloc[:, 8].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=4, metric='minkowski', p=2)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)
```

Compute the confusion matrix.

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
cm = confusion_matrix(y_pred, y_test)
print("The confusion matrix is, \n{}".format(cm))
print("The accuracy is= {}".format(accuracy_score(y_pred, y_test)))
print("The precision is= {}".format(precision_score(y_pred, y_test)))
print("The recall is= {}".format(recall_score(y_pred, y_test)))
print("The f1 score= {}".format(f1_score(y_pred, y_test)))
```

Construct the CAP curve.

```python
total = len(y_test)
class_1_count = np.sum(y_test)
class_0_count = total - class_1_count
plt.figure(figsize=(20,12))
plt.plot([0, total], [0, class_1_count], c = '#0F30E5', linestyle = '--', label = 'Random Model')
plt.plot([0, class_1_count, total], 
         [0, class_1_count, class_1_count], 
         c = '#5AF37F', 
         linewidth = 2, 
         label = 'Perfect Model')
probs = classifier.predict_proba(X_test)
probs = probs[:, 1]
model_y = [y for _, y in sorted(zip(probs, y_test), reverse = True)]
y_values = np.append([0], np.cumsum(model_y))
x_values = np.arange(0, total + 1)
plt.plot(x_values, 
         y_values, 
         c = '#E75130', 
         label = 'K Neighbors Classifier', 
         linewidth = 4)

# Plot information
plt.xlabel('Total observations', fontsize = 16)
plt.ylabel('Is a pulsar star', fontsize = 16)
plt.title('Cumulative Accuracy Profile', fontsize = 16)
plt.legend(loc = 'lower right', fontsize = 16)
plt.show()
```
